```python
df_info = pd.read_csv('M4-info.csv')
df_info.head()
```

```python
df = pd.read_csv('Daily-train.csv')
df.head()
```

```python
df.info()
```

```python
# Join the info file to the train file
df = pd.merge(df, df_info, how='left', left_on='V1', right_on='M4id')
```

```python
df.head()
```

```python
df['category'].unique()
```

```python
# Unique categories
categories = df['category'].unique()

# Splitting the DataFrame and storing in a dictionary
dfs = {f'df_{cat}': df[df['category'] == cat] for cat in categories}

# Accessing each DataFrame
# For example, to access the DataFrame for 'Macro' category
df_macro = dfs['df_Macro']

df_micro = dfs['df_Micro']

df_demographic = dfs['df_Demographic']

df_industry = dfs['df_Industry']

df_finance = dfs['df_Finance']

df_other = dfs['df_Other']
```

D1, D2 etc are days, can specify a datetime range as one would like 

```python
# Set arbitrary date as the first occuring for easier viewing
# Convert column to datetime
start_date = pd.to_datetime('2000-01-01')

# Extract the day number and convert
df['V1'] = df['V1'].str.extract('D(\d+)').astype(int)  # Extract day number and convert to integer
df['Date'] = df['V1'].apply(lambda x: start_date + pd.Timedelta(days=x-1))  # Convert to datetime

df.head()
```

### Forecast

```python
# forecast test 
df_first_30 = df.head(30)
timegpt_fcast = timegpt.forecast(df = df_first_30, h = 5, freq = 'D', time_col= 'Date', target_col = 'V2')
timegpt_fcast.head()
```
